## Carnival Dice Roll

You are at a carnival and come across a person in a booth offering you a game of "chance" (as people in booths at carnivals tend to do).

You pay 5 dollars and roll 3 dice. If the sum of the dice rolls is greater than 12, you get 15 dollars. If it's less than or equal to 12, you get nothing.

Assuming the dice are fair, should you play this game? How would this change if the winning condition was a sum greater than or equal to 12?

```python
n_dice_rolled = 3
target = 12
```

```python
def roll_the_dice(ntrials, n_dice_rolled=3, target=12):
    rolls = np.random.choice([1, 2, 3, 4, 5, 6], (ntrials, n_dice_rolled))
    roll_sums = rolls.sum(axis=1)
    roll_wins = sum([1 for roll in roll_sum if roll > target])
    p_win = roll_wins/ntrials
    win_pctg = 100*p_win
    mean_winnings = (15*p_win) - 5
    print(f'Trials: {ntrials:d}')
    print(f'Winning percentage: {win_pctg:.3f}%')
    print(f'Winnings per roll: ${mean_winnings:.2f}')
    return roll_sums
```

## Nappy Time

There's a 30% chance my son takes a nap on any given weekend day. What is the chance that he takes a nap at least one day this weekend? What is the probability that he doesn't nap at all?

```python
p_nap = .3
naps = np.random.random((10,2)) <= p_nap
```

## Chance of a 3 

```python
def dice_roll(ntrials=10, n_dice=5):
    rolls = pd.DataFrame(np.random.choice([1,2,3,4,5,6], (ntrials, n_dice)))
    rolls['has_3'] = rolls.apply(lambda row: 3 in row.values, axis=1)
    return rolls
```

```python
ntrials = 100000
n_dice = 5
rolls = dice_roll(ntrials=ntrials, n_dice=n_dice)
p_has_3 = sum(rolls.has_3) / ntrials
print(p_has_3)
pd.DataFrame(rolls)
```
